```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt

df=pd.read_excel('CTG.xls',sheet_name='Data',skiprows=1)
```

```python
df.head()
```

```python
df=df.dropna(how='all',axis=1)
```

```python
df.head()
```

```python
df=df.dropna(how='any',axis=0)
```

```python
df.head()
```

```python
df.shape
```

```python
print(df.columns)
```

```python
df.corr()
```

```python
df=df.drop('DR',axis=1)
```

```python
df_features=df.drop('CLASS',axis=1)
df_features=df_features.drop('NSP',axis=1)
```

```python
#train and test spliting data
df_target = df['NSP']
df_target
```

```python
from sklearn.model_selection import train_test_split
df_features_train, df_features_test, df_target_train, df_target_test = train_test_split(df_features,df_target,test_size=0.3,random_state=42)
```

```python
print(df_features_train.shape)
print(df_features_test.shape)
print(df_target_train.shape)
print(df_target_test.shape)

```

```python
#from sklearn.preprocessing import MinMaxScaler
#scalar = MinMaxScaler()
#scalar.fit(df_features_train)
#df_features_train = scalar.transform(df_features_train)
#df_features_test = scalar.transform(df_features_test)
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df_features_train = sc.fit_transform(df_features_train)
df_features_test = sc.transform(df_features_test)
```

```python
#from sklearn.decomposition import PCA
#pca = PCA(n_components=10)
#pca.fit(df_features)
#df_features = pca.transform(df_features)

from sklearn.decomposition import PCA

pca = PCA(n_components=15)
df_features_train = pca.fit_transform(df_features_train)
df_features_test = pca.transform(df_features_test)
```

```python
# using random forest
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(df_features_train, df_target_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
conf = confusion_matrix(df_target_test.values, rf_clf.predict(df_features_test))
print(conf)
label = ["N","S","P"]
plt.figure()
ax = sns.heatmap(conf, annot=True, xticklabels=label, yticklabels=label, cmap="YlGnBu")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()

print(classification_report(df_target_test.values, rf_clf.predict(df_features_test)))
```

```python
#Decision Tree
from sklearn import tree
tr_clf = tree.DecisionTreeClassifier()
tr_clf.fit(df_features_train, df_target_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
conf = confusion_matrix(df_target_test.values, tr_clf.predict(df_features_test))
print(conf)
label = ["N","S","P"]
plt.figure()
ax = sns.heatmap(conf, annot=True, xticklabels=label, yticklabels=label, cmap="YlGnBu")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()

print(classification_report(df_target_test.values, tr_clf.predict(df_features_test)))
```

```python
# Support vector Machine
from sklearn import svm
sv_clf = svm.SVC(kernel='linear')
sv_clf.fit(df_features_train, df_target_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
conf = confusion_matrix(df_target_test.values, sv_clf.predict(df_features_test))
print(conf)
label = ["N","S","P"]
plt.figure()
ax = sns.heatmap(conf, annot=True, xticklabels=label, yticklabels=label, cmap="YlGnBu")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()

print(classification_report(df_target_test.values, sv_clf.predict(df_features_test)))
```
