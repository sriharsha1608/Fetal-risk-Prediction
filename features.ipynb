```python
df_features=df.drop('CLASS',axis=1)
df_features=df_features.drop('NSP',axis=1)
```

```python
df_features.columns
```

# Training and Testing split

```python
df_target = df['NSP']
```

```python
df_target
```

```python
from sklearn.model_selection import train_test_split
df_features_train, df_features_test, df_target_train, df_target_test = train_test_split(df_features,df_target,test_size=0.3,random_state=42)
```

```python
print(df_features_train.shape)
print(df_features_test.shape)
print(df_target_train.shape)
print(df_target_test.shape)
```

```python
#MRMR Feature selection
num = 15 # number of features
filtered = pymrmr.mRMR(df_features,'MIQ', num)
filtered
```

# Training

```python
# using random forest
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(df_features_train[filtered], df_target_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
conf = confusion_matrix(df_target_test.values, rf_clf.predict(df_features_test[filtered]))
print(conf)
label = ["N","S","P"]
plt.figure()
ax = sns.heatmap(conf, annot=True, xticklabels=label, yticklabels=label, cmap="YlGnBu")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()

print(classification_report(df_target_test.values, rf_clf.predict(df_features_test[filtered])))
```
